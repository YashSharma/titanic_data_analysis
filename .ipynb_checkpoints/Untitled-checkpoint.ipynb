```python
import pandas as pd
from pandas import Series,DataFrame

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
tit = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
tit.head()
```

```python
tit = tit.drop(['PassengerId','Name','Ticket'], axis = 1)
test = test.drop(['Name','Ticket'], axis = 1)
```

```python
i=j=q=0
for t in tit['Embarked']:
    if t == 'S':
        i=i+1
    if t == 'C':
        j=j+1
    if t == 'Q':
        q=q+1
        
```

```python
print i
print j
print q
```

```python
tit['Embarked'] = tit['Embarked'].fillna('S')
sns.factorplot('Embarked','Survived', data=tit,size=4,aspect=3)
```

```python
fig, (axis1,axis2,axis3) = plt.subplots(1,3,figsize=(15,5))

sns.countplot(x='Embarked', data=tit, ax=axis1)
sns.countplot(x='Survived', hue="Embarked", data=tit, ax=axis2)

# group by embarked, and get the mean for survived passengers for each value in Embarked
embark_perc = tit[["Embarked", "Survived"]].groupby(['Embarked'],as_index=False).mean()
sns.barplot(x='Embarked', y='Survived', data=embark_perc,order=['S','C','Q'],ax=axis3)
```

```python
plt.xlabel('Age')
plt.ylabel('Frequency')
tit['Age'].hist()
```

```python
facet = sns.FacetGrid(tit, hue='Survived', aspect=4)
facet.map(sns.kdeplot,'Age', shade=True)
facet.add_legend()

fig, axis1 = plt.subplots(1,1,figsize=(15,4))
average_age = tit[["Age", "Survived"]].groupby(['Age'],as_index=False).mean()
plt.xlim(0,50)
sns.barplot(x='Age', y='Survived', data=average_age)
```

```python
# Family
tit['Family'] =  tit["Parch"] + tit["SibSp"]
tit['Family'].loc[tit['Family'] > 0] = 1
tit['Family'].loc[tit['Family'] == 0] = 0

test['Family'] =  test["Parch"] + test["SibSp"]
test['Family'].loc[test['Family'] > 0] = 1
test['Family'].loc[test['Family'] == 0] = 0

# drop Parch & SibSp
tit = tit.drop(['SibSp','Parch'], axis=1)
test = test.drop(['SibSp','Parch'], axis=1)

# plot
fig, (axis1,axis2) = plt.subplots(1,2,sharex=True,figsize=(10,5))

# sns.factorplot('Family',data=titanic_df,kind='count',ax=axis1)
sns.countplot(x='Family', data=tit, order=[1,0], ax=axis1)

# average of survived for those who had/didn't have any family member
family_perc = tit[["Family", "Survived"]].groupby(['Family'],as_index=False).mean()
sns.barplot(x='Family', y='Survived', data=family_perc, order=[1,0], ax=axis2)

axis1.set_xticklabels(["With Family","Alone"], rotation=0)
```

```python
def get_person(passenger):
    age,sex = passenger
    return 'child' if age < 16 else sex
    
tit['Person'] = tit[['Age','Sex']].apply(get_person,axis=1)
test['Person']    = test[['Age','Sex']].apply(get_person,axis=1)

tit.drop(['Sex'],axis=1,inplace=True)
test.drop(['Sex'],axis=1,inplace=True)
```

```python
fig, (axis1,axis2) = plt.subplots(1,2,figsize=(10,5))

# sns.factorplot('Person',data=titanic_df,kind='count',ax=axis1)
sns.countplot(x='Person', data=tit, ax=axis1)

# average of survived for each Person(male, female, or child)
family_perc = tit[["Person", "Survived"]].groupby(['Person'],as_index=False).mean()
sns.barplot(x='Person', y='Survived', data=family_perc, ax=axis2, order=['male','female','child'])

tit.drop(['Person'],axis=1,inplace=True)
test.drop(['Person'],axis=1,inplace=True)
```

```python
sns.factorplot('Pclass','Survived',order=[1,2,3], data=tit,size=5)
```

```python
X_train = tit.drop(['Survived','Embarked','PassengerId','Name','Ticket','Cabin'], axis=1)
Y_train = tit['Survived']
X_test = test.drop(['PassengerId','Name','Ticket','Cabin','Embarked'], axis=1)
```

```python
X_test["Fare"].fillna(X_test["Fare"].median(), inplace=True)
print X_test['Fare'].isnull().sum()
```

```python
tit_avg = X_train['Age'].mean()
tit_std = X_train['Age'].std()
tit_nan = X_train['Age'].isnull().sum()

test_avg = X_test['Age'].mean()
test_std = X_test['Age'].std()
test_nan = X_test['Age'].isnull().sum()

rand_1 = np.random.randint(tit_avg - tit_std, tit_avg + tit_std, size = tit_nan)
rand_2 = np.random.randint(test_avg - test_std, test_avg + test_std, size = test_nan)

# fill NaN values in Age column with random values generated
X_train["Age"][np.isnan(X_train["Age"])] = rand_1
X_test["Age"][np.isnan(X_test["Age"])] = rand_2

# convert from float to int
X_train['Age'] = X_train['Age'].astype(int)
X_test['Age'] = X_test['Age'].astype(int)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
logreg.score(X_train, Y_train)
```

```python
#Support Vector Machines
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
svc.score(X_train, Y_train)
```

```python
# Random Forests
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
knn.score(X_train, Y_train)
```

```python
# Gaussian Naive Bayes
gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
gaussian.score(X_train, Y_train)
```
